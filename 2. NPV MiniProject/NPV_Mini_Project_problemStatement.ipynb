#### E. An Indian male requires 2500 calories in a day and a female requires 2000 calories in a day to maintain good lifestyle. Find the solutions as displayed in sample output1 and 2.

```python
men1
#a= men1.groupby(by='Category_Type')[['Calories','Item_in_Menu']]
#list(a)
#men1.groupby(by='Category_Type')['Calories'].mean()
calq=men1.groupby(by='Category_Type')['Calories'].mean()['Beverages']   #Avg
calq=round(calq,2)
mcpd=2500
wcpd=2000

mcpd1=round(calq/mcpd,2)
wcpd1=round(calq/wcpd,2)
print("* Average calories of 'Beverages' category is ",calq)
print("* On an Average 'Beverages' contributes to men's calories intake is ",mcpd1)
print("* On an Average 'Beverages' contributes to women's calories intake is ",wcpd1)
```
